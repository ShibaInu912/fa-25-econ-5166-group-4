# Data Description
- Background: 本資料呈現各縣市的月份氣溫狀況
- date received: 2025-10-31
- Path to data file: `fa-25-econ-5166-group-4/data/processed`
- Unit of observation: city-month，每一筆資料是「每個縣市在各個月份所測得的氣候資料」
- Sample period: 2002-2024
- Known issues:
    - 有些縣市的測站較小，並沒有提供日照時數（`sunshine_hours`）
    - 新竹縣並沒有合適的測站可供使用，我們的解決辦法是**新竹縣市共用氣象資料**，並在其他data set中將兩縣市合併，以「新竹地區」為單位討論。
    - 有些測站在特定月份會整修，該月份的資料即無法取得。
    - 有些測站經過遷移，可能會造成測量不一致
        - 如：新北市的測站在2023年從板橋遷至新店
- Definition for each variable: 
    - `date`：識別用日期（每個月的第一天）
    - `year`：年份
    - `month`：月份
    - `city_code`：識別用縣市代碼
    - `city`：城市名稱
    - `station`：該縣市提供氣象資料的測站
    - `sunshine_hours`：日照時數
    - `avg_humidity`：月平均相對濕度
    - `avg_temp`：月平均氣溫
    - `rainy_days`：月降水日數
    - `precipitation`：月降水量
